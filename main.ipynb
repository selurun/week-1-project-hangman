```python
import random
from words import words
import string

def get_word(words):
    word = random.choice(words)  # randomly chooses something from the list
    return word

# setup the hidden list
word = list(get_word(words))
hidden = []
for character in word:
    hidden.append("_")
    
#print(hidden)
#print(word)


attempts = 0
max_attempts = 6

# loop until either the player has won or lost
GameOver = False
while not GameOver:
    if attempts < 6:
        #display the attempts remaining, current board, guessed letters
        print(f"You have {max_attempts - attempts} attempts remaining")
    
        #display the attempts remaining, current board, guessed letters
        print(f"The current word is: {' '.join(hidden)}")
    
    #display the attempts remaining, current board, guessed letters
    
    if attempts == 0:
        print("    -------")
        print("    |    | ")
        print("    |    ")
        print("    |   ")
        print("    |    ") 
        print("    |   ")
        print("-----------")    
    if attempts == 1:
        print("    -------")
        print("    |    | ")
        print("    |    0")
        print("    |   ")
        print("    |    ") 
        print("    |   ")
        print("-----------")
        
    if attempts == 2:
        print("    -------")
        print("    |    | ")
        print("    |    0")
        print("    |   / ")
        print("    |    ") 
        print("    |   ")
        print("-----------")
    if attempts == 3:
        print("    -------")
        print("    |    | ")
        print("    |    0")
        print("    |   / \\")
        print("    |    ") 
        print("    |   ")
        print("-----------")
    if attempts == 4:
        print("    -------")
        print("    |    | ")
        print("    |    0")
        print("    |   / \\")
        print("    |    |") 
        print("    |   ")
        print("-----------")
    if attempts == 5:
        print("    -------")
        print("    |    | ")
        print("    |    0")
        print("    |   / \\")
        print("    |    |") 
        print("    |   / ")
        print("-----------") 
    if attempts == 6:
        print("Sorry, you died.")
        print('\n')
        print("    -------")
        print("    |    | ")
        print("    |    0")
        print("    |   / \\")
        print("    |    |") 
        print("    |   / \\")
        print("-----------")
        print(f"The country we are looking for was: {word}")
        GameOver = True
        continue
        
    # ask the player for a character
    letterguessed = input("Please guess a letter: ")

    print('\n\n\n\n')
    
    if letterguessed in hidden:
        print("You already tried this letter and it is in the word. Please try again!")
        continue
        
    if letterguessed in word:
        #if the player guessed correct, show all matched letters and print message
        #loop through and check if letter exist and loop trough hidden and add the letter
        print(f"You guessed correctly! {letterguessed} is in the word")
        for i in range(len(word)):
            character = word[i] #any letter in the random word
            if character == letterguessed:
                hidden[i] = word[i] #hidden of i is equal to word of i 
                word[i] = "_"
    else:
        # if player guessed wrong, print failure message and increment attempts
        print(f"You guessed wrong! {letterguessed} is NOT in the word")
        attempts += 1
        
        
    # if the player has won print a win message and stop looping
    if all("_" == char for char in word):
        print("Congrats, you won!")
        GameOver = True
```
